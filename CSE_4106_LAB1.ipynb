## **openCV**  
OpenCV stands for Open Source Computer Vision Library.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import cv2
from google.colab.patches import cv2_imshow

path = '/content/drive/MyDrive/Colab Notebooks/CSE 4106/image.png'
```

```python
img = cv2.imread(path)
smaller_img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)  # fx and fy are scaling factors
cv2_imshow(smaller_img)
```

```python
# A grayscale image is a type of image that contains only shades of gray, ranging from black (0 intensity) to white (255 intensity).

img = cv2.imread(path, 0)  # converted into gray scale
# img = cv2.imread(path)
# imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
smaller_img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)  # fx and fy are scaling factors
cv2_imshow(smaller_img)
```

```python
# Blurred image

imgBlur = cv2.GaussianBlur(img, (7, 7), 0)
smaller_img = cv2.resize(imgBlur, (0, 0), fx=0.5, fy=0.5)  # fx and fy are scaling factors
cv2_imshow(smaller_img)
```

```python
# Canny is a popular edge detection algorithm used to find the boundaries (edges) in an image.

imgCanny = cv2.Canny(imgBlur, 50, 50)   # threshold value = 100
smaller_img = cv2.resize(imgCanny, (0, 0), fx=0.5, fy=0.5)  # fx and fy are scaling factors
cv2_imshow(smaller_img)
```

```python
import numpy as np
kernel = np.ones((5, 5), np.uint8)
print(kernel)

imgDilation = cv2.dilate(imgCanny, kernel, iterations=3)   # threshold value = 100
smaller_img = cv2.resize(imgDilation, (0, 0), fx=0.5, fy=0.5)  # fx and fy are scaling factors
cv2_imshow(smaller_img)
```

```python
imgErode = cv2.erode(imgDilation, kernel, iterations=3)   # threshold value = 100
smaller_img = cv2.resize(imgErode, (0, 0), fx=0.5, fy=0.5)  # fx and fy are scaling factors
cv2_imshow(smaller_img)
```
